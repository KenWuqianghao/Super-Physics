```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 10,10000)
g = 9.8
m = 2
alpha = 3
delta_t = 0.01

y = [0]
x = [0]
v_x = [0]
v_y = [0]
f_y = [m * g - alpha * v_y[0]]
f_x = [-alpha * v_x[0]]
a_y = [f_y[0] / m]
a_x= [f_x[0] / m]
```

```python
for i in range (1,len(t)):
    y.append(0.5*a_y[i - 1] * delta_t ** 2 + v_y[i - 1] * delta_t + y[i - 1])
    v_y.append(a_y[i - 1] * delta_t + v_y[i - 1])
    f_y.append(m * g - alpha * v_y[i])
    a_y.append(g - alpha * v_y[i] / m)

    x.append(0.5*a_x[i - 1] * delta_t ** 2 + v_x[i - 1] * delta_t + x[i - 1])
    v_x.append(a_x[i - 1] * delta_t + v_x[i - 1])
    f_x.append(- alpha * v_x[i])
    a_x.append(- alpha * v_x[i] / m)
```

```python
df = pd.DataFrame({'Time':t,'Displacement - Y':y,'Velocity - Y ':v_y,'Force - Y':f_y,'Acceleration - Y':a_y, 'Displacement - X':x, 
'Velocity - X':v_x, 'Force - X':f_x, 'Acceleration - X':a_x,})

df
```
